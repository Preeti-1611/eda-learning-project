```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("student_performance.csv")
df.head()
```

```python
df.shape          # rows, columns
```

```python
df.info()         # data types & null values
```

```python
df.describe()     # statistical summary
```

```python
df.isnull().sum()
```

# Univariate Analysis 

```python
plt.hist(df["study_hours"], bins=20)
plt.title("Study Hours Distribution")
plt.xlabel("Study Hours")
plt.ylabel("Frequency")
plt.show()

```

# Categorical Analysis

```python
plt.figure(figsize=(6,4))
sns.barplot(x="gender", y="math_score", data=df)
plt.title("Math Score by Gender")
plt.show()
```

```python
sns.scatterplot(x="study_hours", y="overall_score", data=df)
plt.title("Study Hours vs Overall Score")
plt.show()

```

```python
sns.boxplot(x="gender", y="overall_score", data=df)
plt.title("Overall Score by Gender")
plt.show()

```

```python
sns.boxplot(y=df["overall_score"])
plt.title("Outliers in Overall Score")
plt.show()
```
